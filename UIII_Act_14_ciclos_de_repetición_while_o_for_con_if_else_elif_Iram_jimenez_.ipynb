### Redaccion del Problema : Simula el crecimiento de un capital con interés compuesto durante N años. Muestra el saldo año a año.

```python
# --- SISTEMA DE IA: CONSULTOR FINANCIERO V1.0 ---

# Variables de control de la sesión
sesion_activa = "si"

print("--- IA FINANCIERA: INICIANDO INTERFAZ DE PROYECCIÓN ---")
print("IA: Hola, soy tu asistente de crecimiento patrimonial. Analicemos tu futuro financiero.")

# Ciclo principal interactivo (IA Persona) [1, 2]
while sesion_activa == "si":
    print("\n--- MENÚ DE CONSULTA ---")
    print("1. Iniciar Simulación de Capital")
    print("2. ¿Qué es el Interés Compuesto? (Concepto IA)")
    print("3. Finalizar Sesión")

    opcion = input("\nUsuario, elija una opción (1-3): ") # Captura interactiva [3]

    # Rama 1: Simulación de Capital (Uso de For y Operadores Aritméticos)
    if opcion == "1":
        print("\nIA: Preparando entorno de cálculo...")

        # Entrada de datos (Casting a float e int) [4, 5]
        capital_inicial = float(input("IA: Ingrese el capital inicial (ej. 1000): "))
        tasa_porcentaje = float(input("IA: Ingrese el porcentaje de interés anual (ej. 5): "))
        tiempo_años = int(input("IA: ¿A cuántos años desea proyectar?: "))

        # Conversión de tasa decimal
        tasa_decimal = tasa_porcentaje / 100
        capital_acumulado = capital_inicial

        print(f"\n{'AÑO':<10} | {'SALDO PROYECTADO':<20}")
        print("-" * 35)

        # Ciclo para mostrar saldo año a año [6, 7]
        # Se calcula el crecimiento anual: capital = capital * (1 + tasa) [8]
        for año in range(1, tiempo_años + 1):
            capital_acumulado = capital_acumulado * (1 + tasa_decimal)
            # Formateo de salida con f-strings para alineación [9]
            print(f"Año {año:<7} | ${capital_acumulado:>18.2f}")

        # Veredicto de la IA basado en resultados (Lógica condicional) [10]
        print("-" * 35)
        if capital_acumulado > (capital_inicial * 2):
            print("IA VERDICTO: ¡Excelente inversión! Tu capital se habrá más que duplicado.")
        elif capital_acumulado > capital_inicial:
            print("IA VERDICTO: Crecimiento constante detectado. Tu dinero trabaja para ti.")
        else:
            print("IA VERDICTO: El crecimiento es conservador. Considera ajustar la tasa o el tiempo.")

    # Rama 2: Explicación de Conceptos
    elif opcion == "2":
        print("\nIA EXPLICACIÓN: El interés compuesto es el 'interés sobre el interés'.")
        print("A diferencia del interés simple, aquí las ganancias de cada año se suman")
        print("al capital base para generar nuevas ganancias en el siguiente ciclo.")

    # Rama 3: Salida del sistema
    elif opcion == "3":
        print("\nIA: Guardando reportes... Sesión financiera finalizada. ¡Mucho éxito!")
        sesion_activa = "no" # Rompe la condición del while principal [2]

    # Gestión de errores de entrada en el menú
    else:
        print("\nIA ERROR: Opción no reconocida. Por favor, ingrese un número del 1 al 3.")

print("\n--- SISTEMA APAGADO ---")
print("Programa realizado por Iram Jimenez 0079 NL = 32 con ayuda d ChatGPT y Notebook ")
```
